Resizing the image using pillow library [Image module]

```python
from PIL import Image
```

```python
img = Image.open('/content/img')
img_resize = img.resize((150, 150))
img_resize
```

Save the resized Image

```python
img_resize.save('dog_image_resized.jpg')
```

Displaying the resize image

```python
img_res = mpimg.imread('/content/dog_image_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

Convert RGB to gray_scale using OpenCV

```python
import cv2
```

```python
img = cv2.imread('/content/img')
```

```python
type(img)
```

```python
img.shape
```

```python
gray_scale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(gray_scale_img)
```

```python
gray_scale_img.shape
# (183, 275) --> 3rd value is not mentioned (1)
```
